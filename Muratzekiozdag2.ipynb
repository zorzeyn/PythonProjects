```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"C:\Users\murat\Python Projects\russia_losses_equipment.csv"

# Task 1: Create a function to extract a numpy array of a column
def extract_column_values(filename, column_name):
    data = pd.read_csv(filename)
    return data[column_name].values

# Task 2: Create a numpy array from the "aircraft" column
aircraft_losses = extract_column_values(file_path, "aircraft")

# Task 3: Find daily aircraft losses
daily_losses = np.diff(aircraft_losses, prepend=0)

# Task 4: Find the 3 greatest daily losses of aircrafts
greatest_daily_losses = np.sort(daily_losses)[-3:]

# Task 5: Determine how many aircraft were shot down in summer 2024
data = pd.read_csv(file_path)
data['date'] = pd.to_datetime(data['date'])
summer_2024 = data[(data['date'] >= "2024-06-01") & (data['date'] <= "2024-08-31")]
summer_losses = summer_2024['aircraft'].iloc[-1] - summer_2024['aircraft'].iloc[0]

# Task 6: Find the mean value of aircraft losses in the last 300 days
last_300_days = data.iloc[-300:]
mean_last_300_losses = np.mean(np.diff(last_300_days["aircraft"], prepend=0))

# Task 7: Plot aircraft losses of the first year of the war
first_year = data[data['date'] <= data['date'].iloc[0] + pd.DateOffset(days=365)]
plt.figure(figsize=(10, 6))
plt.plot(first_year['date'], first_year['aircraft'], marker='o')
plt.title("Aircraft Losses in the First Year of the War")
plt.xlabel("Date")
plt.ylabel("Cumulative Aircraft Losses")
plt.grid(True)
plt.show()

# Display results
print("Daily aircraft losses:", daily_losses)
print("3 greatest daily losses of aircrafts:", greatest_daily_losses)
print("Aircraft shot down in summer 2024:", summer_losses)
print("Mean aircraft losses in the last 300 days:", mean_last_300_losses)
```
