## **2. Government**
Greece is a **parliamentary republic**, with three main branches:

- **Executive Branch**:
    - President of the Republic (Head of State)
    - Prime Minister (Head of Government)
- **Legislative Branch**:
    - Unicameral parliament (*Vouli ton Ellinon*) consisting of 300 members.
- **Judicial Branch**:
    - Independent courts including the **Areopagus** and **Council of State**.

## **4. Cultural and Natural Attractions**
Here are some of the most important **cultural** and **natural** attractions in Greece:

### **A. Cultural Attractions**
1. **Acropolis of Athens**:  
   _A UNESCO World Heritage site, home to the Parthenon._
2. **Delphi**:  
   _Ancient sanctuary, home to the oracle of Apollo._
3. **Knossos**:  
   _The largest Bronze Age archaeological site on Crete._

### **B. Natural Attractions**
- **Meteora**:  
  _Monasteries built on top of towering rock formations._
- **Santorini**:  
  _A volcanic island known for its stunning sunsets and blue-domed buildings._
